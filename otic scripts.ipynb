```python
##CODE TO UPLOAD TEXT AND READ IT FROM PDF , PROCESS IT AND FORMART IT 
import PyPDF2
import re

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as pdf_file:# open in binary read mode ('rb')
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text = ''
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

def preprocess_text(text):
    # Remove non-alphanumeric characters and extra whitespaces
    cleaned_text = re.sub(r'[^\w\s\n.,?!]', '', text)
    return cleaned_text

def format_text(text):
    #Add line breaks after periods followed by a space
    formatted_text = re.sub(r'\n+ ', '\n', text)
    return formatted_text

# Provide the path to the uploaded PDF
pdf_path = 'D:/others/classwork/AI/CONCEPT PAPER.pdf'

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Preprocess the extracted text
#preprocessed_text = preprocess_text(pdf_text)


# Format the text for readability
#formatted_text = format_text(preprocessed_text)

print(pdf_text)
```

```python
import nltk
nltk.download('punkt')


from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer

# Preprocessed text from PDF
preprocessed_text = pdf_text   #"""Your preprocessed text goes here..."""

# Initialize the parser and tokenizer
parser = PlaintextParser.from_string(preprocessed_text, Tokenizer("english"))

# Initialize the LexRank summarizer
summarizer = LexRankSummarizer()

# Generate the summary
sentences_count = 5  # Adjust this value to the desired length
summary = summarizer(parser.document, sentences_count)  # You can adjust the sentence count

print("Summary:")
for sentence in summary:
    print(sentence)

# better summarization code needs the use of chart gpt api 
```
